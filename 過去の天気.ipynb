```python
import requests
# https://www.data.jma.go.jp/obd/stats/etrn/view/hourly_s1.php?prec_no=44&block_no=47662&year=2022&month=03&day=01&view=a2
re = requests.get('https://www.data.jma.go.jp/obd/stats/etrn/view/daily_a1.php?prec_no=35&block_no=1039&year=2022&month=3&day=3&view=')
re.encoding = re.apparent_encoding
# print(re.text)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(re.content)
```

```python
ele_table = soup.select("#tablefix1")
if ele_table:
    ele_trs = ele_table[0].select("tr")
    for ele_tr in ele_trs:
        ele_tds = ele_tr.select("td")
        row_data = list(map(lambda x: x.text, ele_tds))
        if row_data:
            print(row_data)
else:
    print('no data')
```
